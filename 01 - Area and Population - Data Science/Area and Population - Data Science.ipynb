# Import modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read the dataset
california_cities=pd.read_csv("california_cities.csv")
# Display the dataset
california_cities
```

```python
# Check the information of the dataset
california_cities.info()
```

```python
# Extract the needed columns

latd = california_cities["latd"]
longd  = california_cities["longd"]
area  = california_cities["area_land_sq_mi"]
population  = california_cities["population_total"]
```

```python
# Plot a scatter plot
fig = plt.figure()
plt.scatter(longd, latd, label=None, c=np.log10(population),
            cmap='viridis', s=area, linewidth=0, alpha=0.5)


plt.xlabel("Longitude",fontsize=16)
plt.ylabel("Latitude",fontsize=16)
plt.title("Illustration of Area and Population in california cities",fontsize=16)
plt.rcParams["figure.figsize"] = (10,7)
plt.colorbar(label="(log10(population))", orientation="vertical")

# now we will craete a legend, we will plot empty lists with the desired size and label
for area in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.3, s=area, label=str(area) + 'km$^2$')
plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='City Areas')

plt.show()

fig. savefig('test.png')
```
