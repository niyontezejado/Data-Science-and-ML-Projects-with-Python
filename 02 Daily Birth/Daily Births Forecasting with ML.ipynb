```python
# Plot
import matplotlib.dates as mdates

date_column = data['date']
births_column = data['births']

plt.figure(figsize=(12, 4))
plt.plot(date_column,births_column, color = 'm')

plt.title("Daily total female births")
plt.xlabel("Date")
plt.ylabel("Number of birth")
plt.show()
```

```python
"""create a Prophet instance to apply the seasonality effects for the task of daily births forecast with Machine Learning"""

from statsmodels.tsa.statespace.tools import diff

fig = plt.figure(figsize=(12,4))
plt.plot(diff(births_column),color='g');
plt.xlabel('Date')
plt.ylabel('Number of Births')
plt.title('Differenced series')
plt.show()
```
